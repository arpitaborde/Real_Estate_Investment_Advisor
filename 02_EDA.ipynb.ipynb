6. Average Price by State

```python
state_price = df.groupby('State')['Price_in_Lakhs'].mean().sort_values()

state_price.plot(
    kind='barh',
    figsize=(10,6)
)
```

7. Average Property Price by City

```python
city_price = df.groupby('City')['Price_in_Lakhs'].mean()

city_price.sort_values(ascending=False).head(15).plot(
    kind='bar',
    figsize=(12,6)
)
```

8. Median Property Age by Locality

```python
locality_age = (
    df.groupby('Locality')['Age_of_Property']
      .median()
      .sort_values(ascending=False)
      .head(20)
)

locality_age.plot(kind='bar')
```
